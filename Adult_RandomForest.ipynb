```python
import pandas as pd
```

```python
df = pd.read_csv("adult.train.txt", header=None)
df
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','cost']
df
```

```python
z = df.drop(['capital-loss'],axis=1)
z
```

```python
x = df.drop(['hours-per-week'],axis=1)
x
```

```python
y = df['hours-per-week'].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_workclass = LabelEncoder()
le_fnlwgt = LabelEncoder()
le_education = LabelEncoder()
le_marital_status = LabelEncoder()
le_occupation = LabelEncoder()
le_relationship = LabelEncoder()
le_race = LabelEncoder()
le_sex = LabelEncoder()
le_native_country = LabelEncoder()
le_cost = LabelEncoder()
```

```python
x['workclass_n'] = le_workclass.fit_transform(x['workclass'])
x['fnlwgt_n'] = le_workclass.fit_transform(x['fnlwgt'])
x['education_n'] = le_workclass.fit_transform(x['education'])
x['marital_status_n'] = le_workclass.fit_transform(x['marital-status'])
x['occupations_n'] = le_workclass.fit_transform(x['occupation'])
x['relationship_n'] = le_workclass.fit_transform(x['relationship'])
x['race_n'] = le_workclass.fit_transform(x['race'])
x['sex_n'] = le_workclass.fit_transform(x['sex'])
x['native_country_n'] = le_workclass.fit_transform(x['native-country'])
x['cost_n'] = le_workclass.fit_transform(x['cost'])
x
```

```python
x_n = x.drop(['workclass','fnlwgt','education','marital-status','occupation','relationship','race','sex','native-country','cost'],axis='columns')
x_n
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_n,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
res_text = model.predict(X_test)
res_text
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, res_text)
MAE
```

```python
from sklearn.metrics import mean_squared_error
RMSE = mean_squared_error(y_test, res_text, squared=False)
RMSE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE = mean_absolute_percentage_error(y_test, res_text)
MAPE
```
