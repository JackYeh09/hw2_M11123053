```python
import pandas as pd
```

```python
df = pd.read_csv("pre-processing.csv")
df
```

```python
x = df.drop(['Rented Bike Count'],axis=1)
x
```

```python
y = df['Rented Bike Count'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
res_text = model.predict(X_test)
res_text
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, res_text)
MAE
```

```python
from sklearn.metrics import mean_squared_error
RMSE = mean_squared_error(y_test, res_text, squared=False)
RMSE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE = mean_absolute_percentage_error(y_test, res_text)
MAPE
```
