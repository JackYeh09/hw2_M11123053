```python
import pandas as pd
```

```python
df_train = pd.read_csv("adult.train.txt", header=None)
df_train
```

```python
df_test = pd.read_csv("adult.test.txt", header=None)
df_test
```

```python
df_train.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','cost']
df_train
```

```python
df_test.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','cost']
df_test
```

```python
z_train= df_train.drop(['capital-loss'],axis=1)
z_train
```

```python
z_test= df_test.drop(['capital-loss'],axis=1)
z_test
```

```python
x_train = z_train.drop(['hours-per-week'],axis=1)
x_train
```

```python
x_test = z_test.drop(['hours-per-week'],axis=1)
x_test
```

```python
y_train = df_train['hours-per-week'].values
y_train
```

```python
y_test = df_test['hours-per-week'].values
y_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_workclass = LabelEncoder()
le_fnlwgt = LabelEncoder()
le_education = LabelEncoder()
le_marital_status = LabelEncoder()
le_occupation = LabelEncoder()
le_relationship = LabelEncoder()
le_race = LabelEncoder()
le_sex = LabelEncoder()
le_native_country = LabelEncoder()
le_cost = LabelEncoder()
```

```python
x_train['workclass_n'] = le_workclass.fit_transform(x_train['workclass'])
x_train['fnlwgt_n'] = le_workclass.fit_transform(x_train['fnlwgt'])
x_train['education_n'] = le_workclass.fit_transform(x_train['education'])
x_train['marital_status_n'] = le_workclass.fit_transform(x_train['marital-status'])
x_train['occupations_n'] = le_workclass.fit_transform(x_train['occupation'])
x_train['relationship_n'] = le_workclass.fit_transform(x_train['relationship'])
x_train['race_n'] = le_workclass.fit_transform(x_train['race'])
x_train['sex_n'] = le_workclass.fit_transform(x_train['sex'])
x_train['native_country_n'] = le_workclass.fit_transform(x_train['native-country'])
x_train['cost_n'] = le_workclass.fit_transform(x_train['cost'])
x_train
```

```python
x_test['workclass_n'] = le_workclass.fit_transform(x_test['workclass'])
x_test['fnlwgt_n'] = le_workclass.fit_transform(x_test['fnlwgt'])
x_test['education_n'] = le_workclass.fit_transform(x_test['education'])
x_test['marital_status_n'] = le_workclass.fit_transform(x_test['marital-status'])
x_test['occupations_n'] = le_workclass.fit_transform(x_test['occupation'])
x_test['relationship_n'] = le_workclass.fit_transform(x_test['relationship'])
x_test['race_n'] = le_workclass.fit_transform(x_test['race'])
x_test['sex_n'] = le_workclass.fit_transform(x_test['sex'])
x_test['native_country_n'] = le_workclass.fit_transform(x_test['native-country'])
x_test['cost_n'] = le_workclass.fit_transform(x_test['cost'])
x_test
```

```python
x_train_n = x_train.drop(['workclass','fnlwgt','education','marital-status','occupation','relationship','race','sex','native-country','cost'],axis='columns')
x_train_n
```

```python
x_test_n = x_test.drop(['workclass','fnlwgt','education','marital-status','occupation','relationship','race','sex','native-country','cost'],axis='columns')
x_test_n
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=1)
model.fit(x_train_n,y_train)
```

```python
res_text = model.predict(x_test_n)
res_text
```

```python
model.score(x_train_n,y_train)
```

```python
model.score(x_test_n,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, res_text)
MAE
```

```python
from sklearn.metrics import mean_squared_error
RMSE = mean_squared_error(y_test, res_text, squared=False)
RMSE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE = mean_absolute_percentage_error(y_test, res_text)
MAPE
```
